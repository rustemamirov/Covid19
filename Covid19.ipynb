# COVID-19 in all countries and Russia

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# Load data
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/'
infected = pd.read_csv(url + 'time_series_covid19_confirmed_global.csv', sep = ',')
dead = pd.read_csv(url + 'time_series_covid19_deaths_global.csv', sep = ',')
cured = pd.read_csv(url + 'time_series_covid19_recovered_global.csv', sep = ',')
```

```python
# Converting dates to the format dd.mm.yy
new_cols = list(infected.columns[:4]) + list(infected.columns[4:].map(lambda x: '{0:02d}.{1:02d}.{2:d}'.format(int(x.split(sep='/')[1]), int(x.split(sep='/')[0]), int(x.split(sep='/')[2]))))
infected.columns = new_cols
cured.columns = new_cols
dead.columns = new_cols
```

```python
# Converting tables to daily data
infected_daily = infected.copy()
infected_daily.iloc[:,4:] = infected_daily.iloc[:,4:].diff(axis=1)
dead_daily = dead.copy()
dead_daily.iloc[:,4:] = dead_daily.iloc[:,4:].diff(axis=1)
cured_daily = cured.copy()
cured_daily.iloc[:,4:] = cured_daily.iloc[:,4:].diff(axis=1)
```

```python
# Determining the last date for which data is available
last_date = infected.columns[-1]
```

```python
# Creating a legend with the number of confirmed cases of infected, cured and dead
l1 = 'Infected at ' + last_date + ' - ' + str(infected.iloc[:, 4:].sum()[-1])
l2 = 'Cured at ' + last_date + ' - ' + str(cured.iloc[:, 4:].sum()[-1])
l3 = 'Dead at ' + last_date + ' - ' + str(dead.iloc[:, 4:].sum()[-1])

# Plotting total graphs for infection, cured and death
fig, ax = plt.subplots(figsize = [16,9])
ax.plot(infected.iloc[:, 4:].sum(), '-', color = 'orange', label = l1)
ax.plot(cured.iloc[:, 4:].sum(), '-', color = 'green', label = l2)
ax.plot(dead.iloc[:, 4:].sum(), '-', color = 'red', label = l3)

ax.legend(loc = 'upper left', prop=dict(size=12))
ax.xaxis.grid(which='minor')
ax.yaxis.grid()
ax.tick_params(axis = 'x', labelrotation = 90)
plt.title('COVID-19 in all countries')
plt.show()
fig.savefig('all_countries_progressive_total')
```
